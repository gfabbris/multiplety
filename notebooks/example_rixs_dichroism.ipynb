```python
import matplotlib.pyplot as plt
import numpy as np
from multiplety.tools import calculate_rixs,generate_polarization,plot_rixs,save_rixs,load_rixs

%matplotlib nbagg
```

## Calculations general parameters

```python
atomic_number = 28
atom_label = 'Ni2+'
gs_config = '2p6 3d8'
es_config = '2p5 3d9'

fdd = 80
fpd = 80
gpd = 80
coreso = 99
valso = 99

symmetry = 'C4h'
crystal_field = True
dq10 = 1.5
d1 = 0.1
d3 = 0.7

magnetic = False
mag_energy = 0.0

final_state_lifetime = 0.05
intermediate_state_lifetime = 0.5

incident_resolution = 0.1
emitted_resolution = 0.1

states_index = [1,2,3]

th = 20
tth = 130

temperature = 300

ein_min = -7.5
ein_max = 1
ein_step = 0.1

eloss_step = 0.02
```

## Calculating and saving LH

```python
save_folder = 'example_rixs_dichroism_files/LH/'
absorption_pol,emission_pol = generate_polarization(th,tth)

rixs_lh = calculate_rixs(atomic_number,atom_label,gs_config,es_config,
                      crystal_field,symmetry,magnetic,fdd,fpd,gpd,coreso,valso,
                      dq10,d1,d3,mag_energy,save_folder,
                      states_index,absorption_pol,emission_pol,
                      final_state_lifetime, intermediate_state_lifetime,
                      incident_resolution,emitted_resolution,temperature=temperature,
                      ein_min = ein_min, ein_max = ein_max, ein_step = 0.1,
                      eloss_min = 1E10, eloss_max = -1E10, eloss_step = eloss_step,
                      verbose = True, skip_calc = False)


save_rixs(rixs_lh,atomic_number,atom_label,gs_config,es_config,
          crystal_field,symmetry,magnetic,fdd,fpd,gpd,coreso,valso,
          dq10,d1,d3,mag_energy,save_folder,
          states_index,absorption_pol,emission_pol,
          final_state_lifetime, intermediate_state_lifetime,
          incident_resolution,emitted_resolution,temperature,
          fname = 'rixs_matrix_lh.dat')
```

## Calculating and saving LV

```python
absorption_pol = [1j/np.sqrt(2),0,-1j/np.sqrt(2)]
save_folder = 'example_rixs_dichroism_files/LV/'

rixs_lv = calculate_rixs(atomic_number,atom_label,gs_config,es_config,
                      crystal_field,symmetry,magnetic,fdd,fpd,gpd,coreso,valso,
                      dq10,d1,d3,mag_energy,save_folder,
                      states_index,absorption_pol,emission_pol,
                      final_state_lifetime, intermediate_state_lifetime,
                      incident_resolution,emitted_resolution,temperature=temperature,
                      ein_min = ein_min, ein_max = ein_max, ein_step = 0.1,
                      eloss_min = 1E10, eloss_max = -1E10, eloss_step = eloss_step,
                      verbose = True, skip_calc = False)

save_rixs(rixs_lv,atomic_number,atom_label,gs_config,es_config,
          crystal_field,symmetry,magnetic,fdd,fpd,gpd,coreso,valso,
          dq10,d1,d3,mag_energy,save_folder,
          states_index,absorption_pol,emission_pol,
          final_state_lifetime, intermediate_state_lifetime,
          incident_resolution,emitted_resolution,temperature,
          fname = 'rixs_matrix_lv.dat')
```

## Plotting

```python
plt.figure()
rixs_plot = plot_rixs(rixs_lh,plt.gca(),cmap = 'jet',zmin = 0,zmax = None)
plt.title('Ni 3d$^8$ LH', fontsize = 16)

plt.figure()
rixs_plot = plot_rixs(rixs_lv,plt.gca(),cmap = 'jet',zmin = 0,zmax = None)
plt.title('Ni 3d$^8$ LV', fontsize = 16)
```
